## Lorenzo Equations
Consider the Lorenzo Equations,
$$ \frac{dx}{dt}=\sigma (y-x), \ \ x(0)=1,$$
$$ \frac{dy}{dt}=x (\rho-z)-y, \ \ y(0)=1,$$
$$ \frac{dz}{dt}=xy-\beta z , \ \ z(0)=1,$$
apply the Euler method
$$w_{i+1}=w_i+hf(t_i,w_i),$$
where $h$ is the stepsize to discretise.

The discrete form of the Lorenzo equations are
$$ x_{i+1}= x_i+h[\sigma (y_i-x_i],\ \ x_0=1$$
$$ y_{i+1}= y_i+h[(x_i) (\rho-z_i)-y_i], \ \ y_0=1$$
$$ y_{i+1}= y_i+h[x_iy_i-\beta z_i] , \ \ z_0=1,$$
where $h=0.25$ is the step size.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from mpl_toolkits.mplot3d import Axes3D
## To output results as a nice table
from IPython.display import HTML

HTML('''<script>
code_show=true; 
function code_toggle() {
 if (code_show){
 $('div.input').hide();
 } else {
 $('div.input').show();
 }
 code_show = !code_show
} 
$( document ).ready(code_toggle);
</script>
<form action="javascript:code_toggle()"><input type="submit" value="Click here to toggle on/off the raw code."></form>''')
```

```python
N=4
t_end=1.0
t_start=0.0
h=((t_end-t_start)/N)

t=np.arange(t_start,t_end+h/2,h)

sigma=8
rho=1
beta=5/3
x=np.ones(5)
y=np.ones(5)
z=np.ones(5)

for i in range(0,4):
    x[i+1]=x[i]+0.25*(sigma * (y[i] - x[i]))
    y[i+1]=y[i]+0.25*((x[i]) * (rho - z[i])- y[i])
    z[i+1]=z[i]+0.25*((x[i]) * (y[i]) - beta * (z[i]))

print("The numerical approximation of Lorenzo Equations:" )    
print('t:', t)
print('x:', x)
print('y:', y)
print('z:', z)
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.gca(projection='3d')
ax.plot(x, y, z)
plt.xlabel('x')
plt.ylabel('y')

plt.title(r'$\sigma$=%f ' %sigma + r' $\rho$=%f' %rho +r' $\beta$=%f' %beta)
plt.show()
```
